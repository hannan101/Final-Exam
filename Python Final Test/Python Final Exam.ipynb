```python
def analyze_file(file_name):
    filename = open(file_name, 'r')
    lines = 0
    words = 0
    characters = 0
    for line in filename:
        wordslist = line.split()
        lines = lines + 1
        words = words + len(wordslist)
        characters = characters + len(line)
    print(lines)
    print(words)
    print(characters)
analyze_file("test.txt")
```

```python
def search_word(fname):
    try:
        text = open(fname, "r") 
          
        # Create an empty dictionary 
        d = dict() 
          
        # Loop through each line of the file 
        for line in text: 
            # Remove the leading spaces and newline character 
            line = line.strip() 
          
            # Convert the characters in line to 
            # lowercase to avoid case mismatch 
            line = line.lower() 
          
            # Split the line into words 
            words = line.split(" ") 
                                 
          
            # Iterate over each word in line 
            for word in words: 
                # Check if the word is already in dictionary 
                if word in d: 
                    # Increment count of word by 1 
                    d[word] = d[word] + 1
                else: 
                    # Add the word to dictionary with count 1 
                    d[word] = 1
          
        # Print the contents of dictionary 
        for key in list(d.keys()): 
            print(key, ":", d[key]) 
    except Exception as e:
        print(e)
search_word("test.txt")
```

```python
# Q. 2 
class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_borrowed = False  

    def borrow(self):
        if not self.is_borrowed:
            self.is_borrowed = True
            print(f"The book '{self.title}' has been borrowed.")
        else:
            print(f"The book '{self.title}' is already borrowed.")

    def return_book(self):
        if self.is_borrowed:
            self.is_borrowed = False
            print(f"The book '{self.title}' has been returned.")
        else:
            print(f"The book '{self.title}' was not borrowed.")

# Example usage
book1 = Book("Computer", "pitern Notern", "123")
book1.borrow()        
book1.return_book()   
```

```python
# Q 4

class User:
    def __init__(self, user_id, name):
        self.user_id = user_id  
        self.name = name        
        self.borrowed_books = [] 
    def borrow_book(self, book):
        """Adds a book to the user's borrowed books list if it's not already borrowed."""
        if not book.is_borrowed:
            book.borrow()
            self.borrowed_books.append(book)
            print(f"{self.name} has borrowed '{book.title}'.")
        else:
            print(f"Sorry, '{book.title}' is currently borrowed by another user.")

    def return_book(self, book):
        """Removes a book from the user's borrowed books list if it was borrowed by the user."""
        if book in self.borrowed_books:
            book.return_book()
            self.borrowed_books.remove(book)
            print(f"{self.name} has returned '{book.title}'.")
        else:
            print(f"{self.name} has not borrowed '{book.title}'.")

user1 = User("U001", "Alice")
book1 = Book("Introduction to business", "dr. John", "33h3")

user1.borrow_book(book1)


user1.return_book(book1)

```

```python
# 5 
class Book:
    def __init__(self, isbn, title, author):
        self.isbn = isbn
        self.title = title
        self.author = author
        self.is_available = True

    def borrow(self):
        if self.is_available:
            self.is_available = False
            print(f"Book '{self.title}' borrowed successfully.")
        else:
            print("Book is not available.")

    def return_book(self):
        self.is_available = True
        print(f"Book '{self.title}' returned successfully.")


class DigitalBook(Book):
    def __init__(self, isbn, title, author, download_link):
        super().__init__(isbn, title, author)
        self.download_link = download_link

    def borrow(self):
        if self.is_available:
            self.is_available = False
            print(f"Digital book '{self.title}' downloaded successfully.")
        else:
            print("Book is not available.")


class AudioBook(Book):
    def __init__(self, isbn, title, author, duration):
        super().__init__(isbn, title, author)
        self.duration = duration

    def borrow(self):
        if self.is_available:
            self.is_available = False
            print(f"Audio book '{self.title}' is now playing.")
        else:
            print("Book is not available.")


class User:
    def __init__(self, user_id, name):
        self.user_id = user_id
        self.name = name
        self.borrowed_books = []

    def borrow_book(self, book):
        if book.is_available:
            book.borrow()
            self.borrowed_books.append(book)
            print(f"{self.name} borrowed {book.title}")
        else:
            print("Book is not available.")

    def return_book(self, book):
        if book in self.borrowed_books:
            book.return_book()
            self.borrowed_books.remove(book)
            print(f"{self.name} returned {book.title}")
        else:
            print("You have not borrowed this book.")


class Library:
    def __init__(self, name):
        self.name = name
        self.books = []
        self.users = []

    def add_book(self, book):
        self.books.append(book)

    def register_user(self, user):
        self.users.append(user)

    def lend_book(self, user_id, isbn):
        for user in self.users:
            if user.user_id == user_id:
                for book in self.books:
                    if book.isbn == isbn:
                        user.borrow_book(book)
                        return
        print("User or book not found.")

    def receive_return(self, user_id, isbn):
        for user in self.users:
            if user.user_id == user_id:
                for book in user.borrowed_books:
                    if book.isbn == isbn:
                        user.return_book(book)
                        return
        print("User or book not found.")
```

```python
# creatig books
book1 = Book("3nj", "John wick", "Elon")
book2 = DigitalBook("ISBN456", "The Hitchhiker's Guide to the Galaxy", "Douglas Adams", "https://example.com/book2")
book3 = AudioBook("ISBN789", "The Silent Patient", "Alex Michaelides", "10 hours")

# creating users
user1 = User(1, "Ali")
user2 = User(2, "subhan")

# creating library
library = Library("Central Library")

#add books and users
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)
library.register_user(user1)
library.register_user(user2)

# test borrowing and returning
library.lend_book(1, "ISBN123")  # user 1 borrows book1
library.lend_book(2, "ISBN456")  # user 2 borrows book2
library.lend_book(1, "ISBN789")  # user 1 borrows book3
library.receive_return(1, "ISBN123")  # User 1 returns book1
library.lend_book(2, "ISBN123")  # user 2 borrows book1 (now available)

# test error handling
library.lend_book(3, "ISBN123")  # user not found
library.lend_book(1, "ISBN999")  # uook not found
```
